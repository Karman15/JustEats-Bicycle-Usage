```python
# reading all files
filenames = glob.glob(os.path.join('datasets', "*.csv"))
cycles = pd.concat((pd.read_csv(file) for file in filenames), ignore_index=True)

# data wrangling
cycles['started_at'] = pd.to_datetime(cycles.started_at, format="%Y-%m-%d %H:%M:%S")
cycles['ended_at'] = pd.to_datetime(cycles.ended_at)

cycles['start_hour'] = pd.to_numeric(cycles['started_at'].apply(lambda x: x.strftime('%H')))
#cycles['start_12_hour_format'] = cycles['started_at'].apply(lambda x: x.strftime('%I%p'))

cycles['start_station_latitude'] = np.radians(cycles['start_station_latitude'])
cycles['end_station_latitude'] = np.radians(cycles['end_station_latitude'])
cycles['start_station_longitude'] = np.radians(cycles['start_station_longitude'])
cycles['end_station_longitude'] = np.radians(cycles['end_station_longitude'])

diff_longitude = cycles['end_station_longitude'] - cycles['start_station_longitude']
diff_latitude = cycles['end_station_latitude'] - cycles['start_station_latitude']

R = 6373.0
a = np.sin(diff_latitude/2)**2 + np.cos(cycles['start_station_latitude']) * np.cos(cycles['end_station_latitude']) * np.sin(diff_longitude/2)**2
c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
displacement = R * c
cycles['displacement'] = displacement

cycles_pre_lockdown = cycles[cycles['started_at'] < '2020-03-23'].reset_index(drop = True)
cycles_post_lockdown = cycles[cycles['started_at'] >= '2020-03-23'].reset_index(drop = True)
cycles_post_1st_lockdown = cycles[('2020-03-23' <= cycles['started_at']) & (cycles['started_at'] < '2020-10-31')].reset_index(drop = True)
cycles_post_2nd_lockdown = cycles['2020-10-31' <= cycles['started_at']].reset_index(drop = True)

cycles_2018 = cycles[cycles['started_at'] < '2019-01-01'].reset_index(drop = True)
cycles_2019 = cycles[('2018-12-31' < cycles['started_at']) & (cycles['started_at'] < '2020-01-01')].reset_index(drop = True)
cycles_2020 = cycles[('2019-12-31' < cycles['started_at']) & (cycles['started_at'] < '2021-01-01')].reset_index(drop = True)
cycles_2021 = cycles['2020-12-31' < cycles['started_at']].reset_index(drop = True)

cycles_2019_summers = cycles[('2019-06-21' <= cycles['started_at']) & (cycles['started_at'] <= '2019-09-23')].reset_index(drop = True)
cycles_2020_summers = cycles[('2020-06-20' <= cycles['started_at']) & (cycles['started_at'] <= '2020-09-22')].reset_index(drop = True)

cycles_2019_winters = cycles[('2019-12-22' <= cycles['started_at']) & (cycles['started_at'] <= '2020-03-20')].reset_index(drop = True)
cycles_2020_winters = cycles[('2020-12-21' <= cycles['started_at']) & (cycles['started_at'] <= '2021-03-20')].reset_index(drop = True)
```

Hypothesis testing for pre_lockdown and post_1st_lockdown months. We perfomed a 2-sample t-test where the vairance is unknown. Null hypothesis: lockdown has no effect on the average duration of bike rides. Alternative hypothesis: lockdown had an effect on the average duration of bike rides.

```python
stats.ttest_ind(cycles_pre_lockdown['duration'], cycles_post_1st_lockdown['duration'], equal_var = False)
```

We can see that the p-value is significantly smaller and hence, we can conclude that there is enough evidence to reject the null hypothesis, i.e., lockdown has no effect on the average duration of bike rides, in favour of the alternative hypothesis.

```python
trip_days_2019 = cycles_2019.copy()
trip_days_2019['start_day'] = trip_days_2019["started_at"].dt.day_name()
weekday = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
trip_days_2019['start_weekday'] = ['weekday' if x in weekday else 'weekend' for x in trip_days_2019['start_day']]
trip_days_2019_weekday = trip_days_2019[trip_days_2019['start_weekday'] == 'weekday']
trip_station_2019 = trip_days_2019_weekday.groupby(["start_hour","start_station_name"]).count()
trip_station_2019 = trip_station_2019['duration'].unstack()
```

```python
trip_station_2019 = trip_station_2019.fillna(0)
trip_station_2019 = trip_station_2019.apply(lambda x: (x-min(x))/(max(x)-min(x)), axis=0)
trip_station_2019_tran = trip_station_2019.transpose()
X = trip_station_2019_tran.to_numpy()
kmeans2019 = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
array_cluster = np.column_stack((trip_station_2019_tran,kmeans2019.labels_))
trip_station_2019_normal = pd.DataFrame(array_cluster)
trip_station_2019_normal_cluster = trip_station_2019_normal.transpose()
trip_station_2019_normal_cluster0 = trip_station_2019_normal_cluster[trip_station_2019_normal_cluster.columns[trip_station_2019_normal_cluster.iloc[-1]==0.000000]]
trip_station_2019_normal_cluster0 = trip_station_2019_normal_cluster0.drop([24])
trip_station_2019_normal_cluster0['average'] = trip_station_2019_normal_cluster0.mean(numeric_only=True, axis=1)

fig, ax = plt.subplots(figsize=(15,7))
trip_station_2019_normal_cluster0.plot(ax=ax,color='purple', alpha=0.05)
trip_station_2019_normal_cluster0["average"].plot(ax=ax,color='purple', alpha=1)
ax.get_legend().remove()
plt.ylabel('Trips number')
plt.xlabel('Hour')
```

```python
trip_days_2020 = cycles_2020.copy()
trip_days_2020['start_day'] = trip_days_2020["started_at"].dt.day_name()
weekday = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
trip_days_2020['start_weekday'] = ['weekday' if x in weekday else 'weekend' for x in trip_days_2020['start_day']]
trip_days_2020_weekday = trip_days_2020[trip_days_2020['start_weekday'] == 'weekday']
trip_station_2020 = trip_days_2020_weekday.groupby(["start_hour","start_station_name"]).count()
trip_station_2020 = trip_station_2020['duration'].unstack()
```

```python
trip_station_2020 = trip_station_2020.fillna(0)
trip_station_2020 = trip_station_2020.apply(lambda x: (x-min(x))/(max(x)-min(x)), axis=0)
trip_station_2020_tran = trip_station_2020.transpose()
X = trip_station_2020_tran.to_numpy()
kmeans2020 = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
array_cluster = np.column_stack((trip_station_2020_tran,kmeans2020.labels_))
trip_station_2020_normal = pd.DataFrame(array_cluster)
trip_station_2020_normal_cluster = trip_station_2020_normal.transpose()
trip_station_2020_normal_cluster0 = trip_station_2020_normal_cluster[trip_station_2020_normal_cluster.columns[trip_station_2020_normal_cluster.iloc[-1]==0.000000]]
trip_station_2020_normal_cluster0 = trip_station_2020_normal_cluster0.drop([24])
trip_station_2020_normal_cluster0['average'] = trip_station_2020_normal_cluster0.mean(numeric_only=True, axis=1)

fig, ax = plt.subplots(figsize=(15,7))
trip_station_2020_normal_cluster0.plot(ax=ax,color='purple', alpha=0.05)
trip_station_2020_normal_cluster0["average"].plot(ax=ax,color='purple', alpha=1)
ax.get_legend().remove()
plt.ylabel('Trips number')
plt.xlabel('Hour')
```

```python
trip_station_2020_normal_cluster1 = trip_station_2020_normal_cluster[trip_station_2020_normal_cluster.columns[trip_station_2020_normal_cluster.iloc[-1]==1.000000]]
trip_station_2020_normal_cluster1 = trip_station_2020_normal_cluster1.drop([24])
trip_station_2020_normal_cluster1['average'] = trip_station_2020_normal_cluster1.mean(numeric_only=True, axis=1)

fig, ax = plt.subplots(figsize=(15,7))
trip_station_2020_normal_cluster1.plot(ax=ax,color='red', alpha=0.05)
trip_station_2020_normal_cluster1["average"].plot(ax=ax,color='red', alpha=1)
ax.get_legend().remove()
plt.ylabel('Trips number')
plt.xlabel('Hour')
```
